# Visualizations

# a.histogram

```python
plt.figure(figsize=(8,8))

plt.subplot(3,2,1)
plt.hist(data['Classification'], color='red',density='true')
plt.title('Histogram Classification')
plt.xlabel('Classifications')
plt.ylabel('Frequency')

plt.subplot(3,2,2)
plt.hist(data['SL'], color='red',density='true')
plt.title('Histogram SL')
plt.xlabel('SL')
plt.ylabel('Frequency')

plt.subplot(3,2,3)
plt.hist(data['SW'], color='red',density='true')
plt.title('Histogram SW')
plt.xlabel('SW')
plt.ylabel('Frequency')

plt.subplot(3,2,4)
plt.hist(data['PL'], color='red',density='true')
plt.title('Histogram PL')
plt.xlabel('PL')
plt.ylabel('Frequency')


plt.subplot(3,2,5)
plt.hist(data['PW'], color='red',density='true')
plt.title('Histogram PW')
plt.xlabel('PW')
plt.ylabel('Frequency')
```

# b.pairplot


```python
sns.pairplot(data)
```

# c. Heat map

```python
sns.heatmap(data.corr(),annot=True)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
```

# d. Box plot

```python
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
sns.boxplot(x='Classification',y='SL',data=data)
plt.subplot(2,2,2)
sns.boxplot(x='Classification',y='SW',data=data)
plt.subplot(2,2,3)
sns.boxplot(x='Classification',y='PL',data=data)
plt.subplot(2,2,4)
sns.boxplot(x='Classification',y='PW',data=data)
```
