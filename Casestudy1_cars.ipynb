# 3. Plotting a bar graph of Male vs Female buyers who participated in the sales

```python
Sales_gender=data.groupby('Buyer Gender').count()
Sales_gender=Sales_gender['Make']
Sales_gender
```

A total of 5052 females and 4948 Males participated in the car sales

```python
Sales_gender.plot(kind='bar',figsize=(12,8))
plt.title('Male Vs Female Buyers', fontsize=22,fontweight="bold")
plt.xlabel('Gender', fontsize=16,fontweight="bold")
plt.ylabel('Sales in Units', fontsize=16,fontweight="bold")
plt.xticks(fontsize=16,rotation=0)
plt.yticks(np.arange(0,5500,250))
plt.grid()
plt.show()
```


A bar graph is plotted above showing Male Vs Female buyers who have participated in the sales and it can be clearly seen that Female buyers are more in number.
