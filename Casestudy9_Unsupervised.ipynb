```python
# performing PCA with 9components
pca = PCA(n_components = 9)
pca.fit(data)
```

```python
data1= pd.DataFrame(scores_pca)
data1.head()
```

# 2. Try out different clustering models in the wine dataset.


# a) KMeans

```python
from sklearn.cluster import KMeans
model_kmeans = KMeans(n_clusters=7)
model_kmeans.fit(data1)
labels = model_kmeans.predict(data1)
print(labels)
```

```python
output = data1.copy()
output["kmeans-cluster"] = labels
output.head()
```

```python
plt.scatter(output[0],output[1], c = output["kmeans-cluster"])
plt.title('Clusters by PCA components')
plt.xlabel('component 1')
plt.ylabel('component2')
plt.show()
```

# b) Hierarchial Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 8, affinity = 'euclidean', linkage = 'ward')
labels = cluster.fit_predict(data1)
labels
```

```python
output1 = data1.copy()
output1["kmeans-cluster"] = labels
output1.head()
```

```python
plt.figure(figsize=(8,10))
plt.scatter(output1[0],output1[1], c = output["kmeans-cluster"])
plt.title('Clusters by PCA components')
plt.xlabel('component 1')
plt.ylabel('component2')
plt.show()
```

In both models we can see overlapping of clusters, this can be avoided by choosing the optimum number of clusters



# 3. Find the optimum number of clusters in each model and create the model with the optimum number of clusters.


# a) KMeans Clustering


```python
wcss = []
for i in range(1,11):
    kmeans = KMeans (n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(data1)
    wcss.append(kmeans.inertia_)
    print(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title("The elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("wcss")
plt.show()  
```

Since there is a bend at n=3, we can take number of clusters as 3

```python
model_kmeans = KMeans(n_clusters=3)
model_kmeans.fit(data1)
labels = model_kmeans.predict(data1)
print(labels)
```

```python
newdata = pd.concat([data.reset_index(drop=True),data1], axis=1)
newdata.columns.values[-9:] = ['component1', 'component2', 'component3', 'component4',
                             'component5', 'component6', 'component7', 'component8', 'component9']
newdata['Cluster label'] = model_kmeans.labels_
newdata.head()
```

```python
newdata['Segment'] = newdata['Cluster label'].map({0:'first',
                                         1:'second',
                                         2:'third'})
```

```python
plt.figure(figsize=(8,8))
x = newdata['component1']
y = newdata['component2']
sns.scatterplot(x,y, hue = newdata['Segment'])
plt.title('Clusters by PCA components')
plt.show()
```

# b) Hierarchial Clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(data1, method = 'ward')
plt.figure(figsize = (15,10))
dendrogram(linked, orientation = 'top',distance_sort='ascending',show_leaf_counts = True)
plt.hlines(20, 0, 5000, linestyles= 'dotted', colors='r')
plt.show()
```

From dendogram also we can see that optimum number of clusters is 3

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
labels = cluster.fit_predict(data1)
labels
```
